```python
from matplotlib import pyplot as plt
```

```python
before_data = []
with open("stats_before", "r") as file:
    for line in file:
        before_data.append(float(line))
```

```python
after_data = []
with open("stats_after", "r") as file:
    for line in file:
        after_data.append(float(line))
```

```python
before_mean = sum(before_data) / len(before_data)
after_mean = sum(after_data) / len(after_data)
coeff = before_mean / after_mean
```

```python
plt.figure(figsize=(20,10))
plt.bar("Before", before_mean, label="Time before opt (-O3)")
plt.bar("After", after_mean, label="Time after opt (-O3)  x" + str(round(coeff, 2)))
plt.ylabel("Time, ms", fontsize=25);
plt.legend(prop={'size': 25})
plt.savefig("diagram.png")
```
